```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img=cv2.imread("/content/6.jpg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gc=cv2.bilateralFilter(gray,11,15,15)
edge=cv2.Canny(gc,170,200)
con,herf=cv2.findContours(edge,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
ctn=sorted(con,key=cv2.contourArea,reverse=True)
for c in ctn:
    peri=cv2.arcLength(c,True)
    epsilon=0.018*peri
    approx=cv2.approxPolyDP(c,epsilon,True)
    if len(approx)==4:
        final=cv2.drawContours(img,(approx),-1,(255,0,0),3)
        break

cv2_imshow(img)
cv2.waitKey(0)    
```
